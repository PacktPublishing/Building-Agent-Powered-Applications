### Sentiment analysis with spaCy

```python
import spacy
nlp = spacy.load("en_core_web_sm")

# Define simple sentiment lexicons
positive_words = {'good', 'great', 'excellent', 'amazing', 'love', 'wonderful', 'happy', 'fantastic', 'liked'}
negative_words = {'bad', 'terrible', 'awful', 'hate', 'horrible', 'worst', 'disappointed', 'sad', 'boring'}

def analyze_sentiment(text):
    doc = nlp(text.lower())

    # Filter tokens: keep lemmatized words, ignore stop-words and punctuation
    tokens = [token.lemma_ for token in doc if token.is_alpha and not token.is_stop]

    # Count sentiment words
    pos_count = sum(1 for word in tokens if word in positive_words)
    neg_count = sum(1 for word in tokens if word in negative_words)

    # Determine sentiment
    if pos_count > neg_count:
        return "Positive"
    elif neg_count > pos_count:
        return "Negative"
    else:
        return "Neutral"

# Example usage
text = "The movie was amazing! I loved every minute of it."
sentiment = analyze_sentiment(text)
print(f"Sentiment: {sentiment}")
```

### Text classification with Naive Bias model

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB

# Small labeled dataset
texts = [
    "I love this movie!",
    "Absolutely fantastic experience!",
    "Worst film I’ve ever seen.",
    "Terrible acting and boring plot.",
    "Amazing visuals and great soundtrack",
    "Disappointing and dull"
]
labels = ['positive', 'positive', 'negative', 'negative', 'positive', 'negative']

# Step 1: Convert text to feature vectors
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(texts)

# Step 2: Train the Naive Bayes model
model = MultinomialNB()
model.fit(X, labels)

# Step 3: Test on a new example
new_text = ["The story was boring but the visuals were amazing"]
X_new = vectorizer.transform(new_text)
prediction = model.predict(X_new)

print(f"Sentiment: {prediction[0]}")
```
