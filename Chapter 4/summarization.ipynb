### Filter similar paragraphs for text summarization

```python
from sentence_transformers import SentenceTransformer, util

# Sample paragraphs
paragraphs = [
    "AI is transforming industries through automation and data analysis.",
    "Automation and machine learning are changing how industries operate.",
    "Recent developments in AI have introduced new capabilities in image and speech recognition.",
    "Industries are being revolutionized by AI through better decision-making and automation."
]

# Load a sentence transformer model
model = SentenceTransformer('all-MiniLM-L6-v2')

# Compute embeddings for each paragraph
embeddings = model.encode(paragraphs, convert_to_tensor=True)

# Set a similarity threshold and initialize selected paragraphs
similarity_threshold = 0.8
selected_indices = [0]

# Compare each paragraph to already selected ones
for idx in range(1, len(paragraphs)):
    is_similar = False
    for selected_idx in selected_indices:
        similarity = util.cos_sim(embeddings[idx], embeddings[selected_idx]).item()
        if similarity > similarity_threshold:
            is_similar = True
            break
    if not is_similar:
        selected_indices.append(idx)

# Final filtered paragraphs
filtered_paragraphs = [paragraphs[i] for i in selected_indices]

print("Filtered Paragraphs:")
for para in filtered_paragraphs:
    print("-", para)
```
