### Define A2A server

```python
from google.adk import Agent
from google.adk.a2a.utils.agent_to_a2a import to_a2a

# Define the agent
root_agent = Agent(
    model="gemini-2.0-flash",
    name="hello_world_agent",
    instructions="You are a helpful assistant that responds with 'Hello, World!' to any input"
)

# Expose the agent via A2A
a2a_app = to_a2a(root_agent, port=8001)

# Start the HTTP server (run with uvicorn):
# uvicorn path.to.module:a2a_app --host localhost --port 8001
```
