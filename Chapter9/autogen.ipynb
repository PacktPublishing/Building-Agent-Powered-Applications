### AutoGen agents

```python
from autogen.agentchat import AssistantAgent, GroupChat, GroupChatManager

llm_config = {
    "cache_seed": 42, "temperature": 0.5,
    "config_list": [ { "provider": "openai", "model": "gpt-4o", "api_key": "<OPENAI_API_KEY>"}],
}

# define two LLM-powered agents
coder = AssistantAgent(
    name="Coder",
    system_message=("You are a senior software engineer. Generate clean, production-quality Python code"),
    llm_config=llm_config,
)
reviewer = AssistantAgent(
    name="Reviewer",
    system_message=("You are a code reviewer. Review code for correctness, performance, and tests."),
    llm_config=llm_config,
)

# put agents in a group chat and let a manager route the conversation
group_chat = GroupChat(agents=[coder, reviewer], max_round=4, speaker_selection_method="auto")
manager = GroupChatManager(groupchat=group_chat, llm_config=llm_config)

# start the conversation with a task for the Coder
task = (
    "Write a Python function `slugify(text: str) -> str` that converts arbitrary text "
    "into a URL-friendly slug (lowercase, hyphen-separated, strip punctuation/accents). "
    "Also include a small pytest file with 3–4 tests."
)
result = manager.initiate_chat(coder, message=task)

# print results
print("\n--- Final Message ---\n")
print(result.summary or result.chat_history[-1]["content"])
```
