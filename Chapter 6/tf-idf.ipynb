### TF-IDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd

# Step 1: Collect Documents (Corpus)
corpus = [
    "Invoice payment is pending",
    "Customer made a payment",
    "Invoice was processed for customer",
    "Payment is completed",
    "Customer has pending invoices"
]

# Step 2–5: Preprocessing + TF-IDF Vectorization
vectorizer = TfidfVectorizer(lowercase=True, stop_words="english")
tfidf_matrix = vectorizer.fit_transform(corpus)

# Get feature names (words)
feature_names = vectorizer.get_feature_names_out()

# Convert TF-IDF matrix to DataFrame for readability
df_tfidf = pd.DataFrame(tfidf_matrix.toarray(), columns=feature_names, index=[f"Doc {i+1}" for i in range(len(corpus))])

# Display the TF-IDF table
print(df_tfidf)
```

### Search by keyword

```python
# Define the keyword you want to search for
keyword = "invoice"

# Check if the keyword is in the vocabulary
if keyword in feature_names:
    # Get the column index of the keyword
    keyword_index = list(feature_names).index(keyword)

    # Find all document indices where the TF-IDF value is greater than 0
    matching_docs = df_tfidf[df_tfidf[keyword] > 0]

    print(f"\nDocuments containing the word '{keyword}':\n")
    print(matching_docs)
else:
    print(f"\nThe word '{keyword}' is not in the vocabulary.")
```
