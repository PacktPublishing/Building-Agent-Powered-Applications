### Semantic search with FAISS

```python
from sentence_transformers import SentenceTransformer
import faiss

# Step 1: Sample documents
documents = [
    "The invoice has been paid.",
    "Customer requested a refund.",
    "Payment is still pending.",
    "The shipment was delivered yesterday.",
    "Refund has been processed."
]

# Step 2: Generate embeddings using a transformer model
model = SentenceTransformer('all-MiniLM-L6-v2')
doc_embeddings = model.encode(documents, convert_to_numpy=True)

# Step 3: Build FAISS index
dimension = doc_embeddings.shape[1]  # e.g., 384 for MiniLM
index = faiss.IndexFlatL2(dimension)  # L2 similarity (can also use cosine)
index.add(doc_embeddings)

# Step 4: Encode query
query = "Has the customer paid the invoice?"
query_embedding = model.encode([query], convert_to_numpy=True)

# Step 5: Perform similarity search
k = 2  # top-k results
distances, indices = index.search(query_embedding, k)

# Step 6: Show results
print("Query:", query)
print("\nTop matching documents:")
for rank, idx in enumerate(indices[0]):
    print(f"{rank + 1}. {documents[idx]} (distance: {distances[0][rank]:.4f})")
```
