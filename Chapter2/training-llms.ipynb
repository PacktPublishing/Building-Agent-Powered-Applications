### Use GPU to perform matrix operations

```python
import torch

# Check if CUDA is available
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# Create two random matrices and move them to the GPU
a = torch.randn(3000, 3000).to(device)
b = torch.randn(3000, 3000).to(device)

# Perform matrix multiplication on the GPU
result = torch.matmul(a, b)

print(result)
```
