### Controlling Output Length with the OpenAI ChatCompletion API.

```python
import openai

openai.api_key = "your-api-key"

# Example of setting a maximum number of output tokens
response = openai.ChatCompletion.create(
    model="gpt-4o",
    messages=[
        {"role": "user", "content": "Summarize the history of artificial intelligence in 100 words."}
    ],
    max_tokens=200,  # Limit the maximum output tokens
    temperature=0.7
)

print(response['choices'][0]['message']['content'])
```
