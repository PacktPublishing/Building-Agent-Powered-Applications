### System and user prompt for GPT models

```python
import openai
openai.api_key = "your-api-key"

response = openai.ChatCompletion.create(
    model="gpt-4",
    messages=[
        {
            "role": "system",
            "content": "You are a friendly customer support assistant for a software company. ... "
        },
        {
            "role": "user",
            "content": "How can I reset my password?"
        }
    ]
)

print(response["choices"][0]["message"]["content"])
```

### System and user prompt for Anthropic's Claude LLMs

```python
import anthropic

client = anthropic.Anthropic(api_key="your-anthropic-api-key")

response = client.messages.create(
    model="claude-3-sonnet-20240229",
    max_tokens=1000,
    messages=[
        {
            "role": "system",
            "content": "You are a friendly customer support assistant for a software company. ... "
        },
        {
            "role": "user",
            "content": "How can I reset my password?"
        }
    ]
)

print(response.content[0].text)
```

### Send completion request to LLaMa model

```python
from transformers import AutoTokenizer, AutoModelForCausalLM

# Load model and tokenizer (ensure you have access)
model_name = "meta-llama/Llama-2-7b-chat-hf"  # Or another LLaMA-based model
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)

# Construct a single prompt combining system and user roles
prompt = (
    "You are a friendly customer support assistant. ... \n" # system prompt
    "User: How can I reset my password?\n"                  # user prompt
)

# Tokenize input
inputs = tokenizer(prompt, return_tensors="pt")

# Generate response
outputs = model.generate(**inputs, max_new_tokens=150, do_sample=True, temperature=0.7, top_p=0.9)

# Decode and print the response
response = tokenizer.decode(outputs[0], skip_special_tokens=True)
print(response)
```
