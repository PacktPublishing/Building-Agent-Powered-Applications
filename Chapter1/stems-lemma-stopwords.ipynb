### Difference Between Stemming and Lemmatization in Python (spaCy & NLTK).

```python
import spacy
from nltk.stem import PorterStemmer

nlp = spacy.load("en_core_web_sm")
stemmer = PorterStemmer()

# Sample words
words = ["running", "fairly", "better", "flies", "wolves"]

# Process with spaCy
doc = nlp(" ".join(words))

# Apply stemming and lemmatization
for token in doc:
    print(f"Word: {token.text:<10} | Stem: {stemmer.stem(token.text):<10} | Lemma: {token.lemma_}")
```

### Retrieving Stop Words Using spaCy.

```python
import spacy

# Load spaCy English language model
nlp = spacy.load("en_core_web_sm")

# Sample text
text = "This is a simple sentence that contains some stop words."

# Process the text
doc = nlp(text)

# Print stop words
print("Stop words in the text:")
for token in doc:
    if token.is_stop:
        print("-", token.text)
```
