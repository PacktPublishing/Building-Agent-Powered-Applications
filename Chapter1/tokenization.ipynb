### Tokenizing Text with the BERT Tokenizer (Hugging Face Transformers).

```python
# using Hugging Face transformers library
from transformers import AutoTokenizer

# Load a tokenizer from BERT model
tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")

sentence = "Tell me a joke"

# tokenizing the sentence
tokens = tokenizer.tokenize(sentence)

# Get token IDs
token_ids = tokenizer.convert_tokens_to_ids(tokens)

# Output results
print("Tokens:", tokens)
print("Token IDs:", token_ids)
```

### Tokenizing Out-of-Vocabulary Words with the T5 Tokenizer (Hugging Face Transformers).

```python
from transformers import AutoTokenizer

# Load a tokenizer from the T5 model
tokenizer = AutoTokenizer.from_pretrained("t5-small")

# Input sentence
sentence = "Microsoft Copilot"

# Tokenize the sentence
tokens = tokenizer.tokenize(sentence)
token_ids = tokenizer.convert_tokens_to_ids(tokens)

# Output results
print("Tokens:", tokens)
print("Token IDs:", token_ids)
```

### Sentence SeTokenization with spaCy.

```python
import spacy

# Load spaCy English language model
nlp = spacy.load("en_core_web_sm")

# Sample text
text = "Dr. Smith arrived at 3 p.m. He was tired. However, he continued working. Amazing, isn't it?"

# Process the text with spaCy
doc = nlp(text)

# Iterate through detected sentences
print("Detected sentences:")
for sent in doc.sents:
    print("-", sent.text)
```
