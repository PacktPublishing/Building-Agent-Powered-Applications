### BLUE, ROUGE and METEOR metrics

```python
import sacrebleu
from rouge_score import rouge_scorer
from nltk.translate.meteor_score import meteor_score
import nltk
nltk.download('wordnet', quiet=True)  # for METEOR synonyms/stemming

reference = "The conference focused on advances in renewable energy, featuring talks on solar, wind, and battery technologies. Industry leaders discussed challenges in scaling and integration."
candidates = [
    "Speakers at the conference discussed renewable energy developments, including solar, wind, and batteries, and addressed integration challenges.",
    "A conference was held to explore sustainability, highlighting climate change and the need for cleaner technologies."
]

# BLEU (sacreBLEU expects list of refs; can pass multiple refs)
for i, cand in enumerate(candidates, 1):
    bleu = sacrebleu.corpus_bleu([cand], [[reference]]).score
    print(f"\nCandidate {i} — sacreBLEU: {bleu:.2f}")

# ROUGE (rouge-score)
scorer = rouge_scorer.RougeScorer(['rouge1','rouge2','rougeL'], use_stemmer=True)
for i, cand in enumerate(candidates, 1):
    scores = scorer.score(reference, cand)
    print(f"\nCandidate {i} — ROUGE:")
    print("ROUGE-1 F:", scores['rouge1'].fmeasure)
    print("ROUGE-2 F:", scores['rouge2'].fmeasure)
    print("ROUGE-L F:", scores['rougeL'].fmeasure)

# METEOR (NLTK)
for i, cand in enumerate(candidates, 1):
    m = meteor_score([reference.split()], cand.split())
    print(f"\nCandidate {i} — METEOR: {m:.4f}")
```

### BERTScore

```python
from bert_score import score

# Define the reference summary (golden output) and 2 candidates
reference = [
    "The conference focused on advances in renewable energy, featuring talks on solar, wind, and battery technologies. Industry leaders discussed challenges in scaling and integration."
]
candidates = [
    "Speakers at the conference discussed renewable energy developments, including solar, wind, and batteries, and addressed integration challenges.",
    "A conference was held to explore sustainability, highlighting climate change and the need for cleaner technologies."
]

# Compute BERTScore
P, R, F1 = score(candidates, reference * len(candidates), lang="en", verbose=True)

# Display results
for i, candidate in enumerate(candidates):
    print(f"\nCandidate {i+1}: {candidate}")
    print(f"Precision: {P[i].item():.4f}")
    print(f"Recall:    {R[i].item():.4f}")
    print(f"F1 Score:  {F1[i].item():.4f}")
```

### Sentence embeddings and Cosine similarity

```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity

# Load a pretrained sentence embedding model
model = SentenceTransformer('all-MiniLM-L6-v2')  # Fast and lightweight

# Define reference and candidate summaries
reference = "The conference focused on advances in renewable energy, featuring talks on solar, wind, and battery technologies. Industry leaders discussed challenges in scaling and integration."

candidates = [
    "Speakers at the conference discussed renewable energy developments, including solar, wind, and batteries, and addressed integration challenges.",
    "A conference was held to explore sustainability, highlighting climate change and the need for cleaner technologies."
]

# Compute embeddings
reference_embedding = model.encode([reference])
candidate_embeddings = model.encode(candidates)

# Compute cosine similarity
cosine_scores = cosine_similarity(candidate_embeddings, reference_embedding)

# Display results
for i, score in enumerate(cosine_scores):
    print(f"Candidate {i+1} Cosine Similarity: {score[0]:.4f}")
```
